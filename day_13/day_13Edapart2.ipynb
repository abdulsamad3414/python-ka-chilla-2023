```python
# visualize and explore
sns.countplot(data=df,x='survived')  # count plot only work on x only
```

```python
sns.barplot(data=df,x='survived',y='fare',hue='who') #in the barplot we will one caterogic and other numeric
# on x-axies we will kept categoric and y-axais we kept numeric
```

```python
sns.countplot(data=df,x='pclass',hue='who')
```

```python
sns.histplot(data=df,x='fare')
```

```python
sns.histplot(data=df,x='age')
```

```python
# use bins
sns.histplot(data=df,x='age',bins=(35))  #via bins we can concluded more clarification
```

```python
sns.histplot(data=df,x='age',color='darkred', alpha=0.7) # here alpha is the color contranst like light or dark
# histogram plot is always making of numeric data not a categoric data
```

```python
sns.boxplot(data=df,x='survived',y='age') #in boxplot() this we will consider 2 values one is categoric and other is numeric
```

```python
#further classification of above cell boxplot
sns.boxplot(data=df,x='survived',y='age',hue='who')
```

```python
#numeric data plot always on scatterplot()
sns.scatterplot(data=df,x='age',y='fare')
```

```python
# we neglect the outliers through the scatterplot by using the lim()
sns.scatterplot(data=df,x='age',y='fare')
plt.ylim(0,300)
```

```python
#further classification
sns.scatterplot(data=df,x='age',y='fare',hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(df,hue='survived')
```

```python
#swarmplot
sns.swarmplot(data=df,x='survived',y='fare')
```

```python
# further classifications by hue function
sns.swarmplot(data=df,x='survived',y='fare',hue='pclass')
```

```python
#for colors
sns.swarmplot(data=df,x='survived',y='fare',hue='who')
```

```python
sns.catplot(data=df,x='pclass',y='fare',hue='embarked',kind='point') # if we write the kind='point' so plot will become a line plot
```

```python
#relationship between two numeric variable
sns.lmplot(data=df,x='age',y='fare')
```

```python
#remove outliers
sns.lmplot(data=df,x='age',y='fare')
plt.ylim(0,250)
```

```python
#further classification
sns.lmplot(data=df, x='age', y='fare', hue='pclass')
plt.ylim(0, 250)  # Set the y-axis limit
# plt.show()
```

```python
#how to add a column and make a age group
sns.histplot(data=df,x='age')
```

```python
#we added age_group column,cut and named it as per desire.
df['age_group']=pd.cut(df['age'],[0,10,18,30,50,80],labels=['kids','child','youth','middle_age','old'])
df.head()
```

```python
df.groupby('age_group')['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(df,hue='survived',vars=['age','fare','pclass'])
```

```python
# we added another column in it
df['family_size']= df['sibsp'] + df['parch']
df
```

```python
df.groupby('family_size')['fare'].mean().plot(kind='bar')
```
