# OUTLIERS REMOVAL

VISUALIZATION WALA METHOD

```python
import seaborn as sns
sns.boxplot(data=df,x='age', y='sex')
```

its mean that more than 55 years old all are outliers in my data

```python
#IQR Method
Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df=df[(df['age'] > lower_bound) & (df['age'] < upper_bound)]
```

```python
df.shape
```

in above method we removed outliers from age because out total rows will 891 and now 825 means 66 are outliers

```python
df.shape
```

```python
# now we will remove fare's outliers
Q1=df['fare'].quantile(0.25)
Q3=df['fare'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
df=df[(df['fare']>lower_bound)& (df['fare']<upper_bound)]
```

```python
df.shape
```

```python
sns.histplot(data =df,x='age')
```

not compltely normalize but outliers removed

```python
sns.histplot(data=df,x='fare')
```

```python
# z-score
from scipy import stats
import numpy as np
df['zscore']=(df['age'] - df['age'].mean()) / df['age'].std()

# zscore=np.abs(stats.zscore(df['age']))
threshold=3
df=df[df['zscore'].abs() <= threshold]
df
# df=df[(z<threshold).all(axis=1)]
```

# Log transformation (jis say data automically scaling leta ha )

```python
# for log transformation we will consider data from start and then working on it
kashti=sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
kashti.head()
```

```python
import numpy as np

kashti['age']=kashti['age'].fillna(value=kashti['age'].median())
kashti['fare']=kashti['fare'].fillna(value=kashti['fare'].median)

#log transform
kashti['age']=np.log(kashti['age']) # if age is children like 2 years 3 years 6 month so we will take log1p
kashti['fare']=np.log(kashti['fare'])
kashti.head()
```

```python
#boxplot after applying the log transformation
sns.boxplot(data=kashti,x='sex',y='age')
```

```python
#By finding the outliers by using the boxplot() after applying log transformation
sns.boxplot(data=kashti,x='sex',y='fare')
```
