# 16-spiltting string into multiple columns

```python
df=pd.DataFrame({'name':['Abdul Samad','Muhmmad fahad','asad ullah'],'Location' :['Pakistan karachi','karachi sharafi','lalokhat pakistan']})
df
```

```python
df.name.str.split(' ').head() # we can split this string into a list of words 
```

```python
# spliting a column into multiple columns
df.name.str.split(' ',expand=True)
```

```python
df['first_name','last_name']=df.name.str.split(' ') # not use expand function 
df
```

```python
df[['first_name','last_name']]=df.name.str.split(' ',expand=True) # use expand function  by first and last name 
df
```

```python
df[['country_name','city_name']]=df.Location.str.split(' ',expand=True)
df
```
