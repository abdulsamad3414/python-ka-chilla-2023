how to deal missing values in EDA analysis?

categorical,numerical(float,int),boolean

```python
print(df2['age'].min())
print(df2['age'].max())
print(df2['age'].mean())
```

```python
df2['age'].unique()
```

```python
# finding the max value in age numerical column
df2['age'].max()
```

```python
#note :here inplace=true is most inportant other wise data will not update
df2['age'].fillna(df2['age'].mean(),inplace=True)
df2
```

```python
df3=df2
```

```python
df3.isnull().sum()
```

```python
#remove deck column in our datset
# df3=df3.drop(columns=['deck'])
```

```python
df3
```

```python
# we will check deck column removed ?
df3.isnull().sum()
```

```python
df4=df3   # make a copy for further analysis
```

```python
df4
```

```python
df4['embark_town'].fillna(df4['embark_town'].mode()[0],inplace=True)
df4
```

```python
df4.isnull().sum()
```

```python
# this is categorical/object column this will be deal like below code and mode is must use in it
df4['embarked'].fillna(df4['embarked'].mode()[0],inplace=True)
df4
```

```python
# finding the mean ,median and mode
print ('the mean is',df['age'].mean())
print('The median is ',df['age'].median())
print('The mode is ',df['age'].mode())  # this is numeric column but usually mode based on categorical data
```

```python
# finding mode on categoric columns
print ('The mode of categoric column is ',df['embark_town'].mode()[0])   #by defult indexing will be [0]if i am not mentioning index so it will take itself
```

```python
#find mode on another categoric columns
print('The mode of categoric column is ',df['embarked'].mode())  #here not mentioned any index
```
